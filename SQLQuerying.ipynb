## Step 7: Data Visualization

```python
# Create a countplot
ax = sns.countplot(x='pet_has_active_subscription', data=pet)

# Add count values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

sns.countplot(x='pet_has_active_subscription', data=pet)
plt.title('Subscription Status of Pets')
plt.xlabel('Subscription Status')
plt.ylabel('Count of Pets')
plt.show()
```

```python
gender_distribution = pet['gender'].value_counts()
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution of Pets')
plt.show()
```

```python
pet['order_payment_date'] = pd.to_datetime(pet['order_payment_date'])
pet.set_index('order_payment_date', inplace=True)
pet.resample('M').size().plot()
plt.title('Monthly Order Payment Trend')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Choose relevant numerical columns for the heatmap
numerical_columns = [
    'pet_order_number', 'wet_food_order_number', 'orders_since_first_wet_trays_order',
    'total_order_kcal', 'wet_trays', 'total_web_sessions', 'total_minutes_on_website'
]

# Create a pivot table
pivot_table = pd.pivot_table(pet, values=numerical_columns, index='pet_has_active_subscription', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Columns')
plt.xlabel('Numerical Columns')
plt.ylabel('Pet Has Active Subscription')
plt.show()
```
